```python
from binance import *
import pandas as pd
import numpy as np
import csv
from datetime import datetime
```

```python
client = Client()
raw_data = client.get_historical_klines('ETHUSDT', client.KLINE_INTERVAL_1HOUR, '1-Apr-2022', '30-Apr-2022')
raw_data
```

```python
data = 'ETH_1H.csv'
target = open(data, 'w', newline='')
candle_writer = csv.writer(target, delimiter=',')
```

```python
for i in raw_data:
  candle_writer.writerow(i)
target.close()
```

```python
csv_data = np.genfromtxt(data, delimiter=',')
csv_data
```

```python
df = pd.DataFrame(csv_data[:1000, 0:6], columns=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume'])
df.head()
```

```python
df.shape
```

```python
df['Index'] = pd.to_datetime(df['Timestamp'],unit='ms')
df.tail()
```

```python
df = df.set_index('Index')
df
```
